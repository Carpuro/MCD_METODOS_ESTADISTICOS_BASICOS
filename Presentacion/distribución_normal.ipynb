# Distribuci贸n de Puntuaciones de CI

El Coeficiente Intelectual (CI) es una puntuaci贸n estandarizada dise帽ada para que los resultados de una poblaci贸n grande sigan una distribuci贸n normal. Por definici贸n, la puntuaci贸n media (渭) del CI es 100 y la desviaci贸n est谩ndar () es 15 puntos.

Vamos a simular los resultados de CI de un grupo de 1,500 personas para ver c贸mo se distribuyen.

$$Z = \frac{x - \mu}{\sigma}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import seaborn as sns

sns.set_style('whitegrid')
```

```python
# Par谩metros de la distribuci贸n de CI
media_ci = 100
desviacion_ci = 15
n_personas = 1500

# Generamos las puntuaciones de CI simuladas
puntuaciones_ci = np.random.normal(loc=media_ci, scale=desviacion_ci, size=n_personas)

# Mostramos las primeras 10 puntuaciones generadas
print("Primeras 10 puntuaciones de CI:", puntuaciones_ci[:10])
```

```python
# Crear el histograma de las puntuaciones de CI
plt.figure(figsize=(10, 6))
plt.hist(puntuaciones_ci, bins=30, density=True, alpha=0.7, color='indigo', label='Histograma de CI')

# A帽adir t铆tulos y etiquetas
plt.title('Distribuci贸n de Puntuaciones de CI (n=1500)')
plt.xlabel('Puntuaci贸n de CI')
plt.ylabel('Densidad')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 7))

# Histograma de los datos
plt.hist(puntuaciones_ci, bins=30, density=True, alpha=0.7, color='indigo', label='Histograma de CI')

# Calculamos la media y desviaci贸n de nuestros datos para la curva
media_calculada = np.mean(puntuaciones_ci)
desviacion_calculada = np.std(puntuaciones_ci)

# Creamos el rango de valores para la curva te贸rica
xmin, xmax = plt.xlim()
x_axis = np.linspace(xmin, xmax, 100)
pdf = norm.pdf(x_axis, media_calculada, desviacion_calculada)

# Dibujamos la curva normal
plt.plot(x_axis, pdf, 'k', linewidth=2, color='lime', label='Curva Normal Te贸rica')

# A帽adimos una l铆nea para la media
plt.axvline(media_calculada, color='red', linestyle='dashed', linewidth=2, label=f'Media: {media_calculada:.2f}')

# T铆tulos y leyenda
plt.title('Distribuci贸n de Puntuaciones de CI con Curva Normal')
plt.xlabel('Puntuaci贸n de CI')
plt.ylabel('Densidad')
plt.legend()
plt.show()

print(f"Media calculada: {media_calculada:.2f}")
print(f"Desviaci贸n est谩ndar calculada: {desviacion_calculada:.2f}")
```

驴Cu谩l es la probabilidad de seleccionar al azar a una persona con un CI inferior a 115?

```python
# Funci贸n para graficar la distribuci贸n y el 谩rea de probabilidad
def graficar_normal(media_calculada, desviacion_calculada, x_min_sombreado, x_max_sombreado, titulo):
    # Crear el rango de valores para el eje X
    x = np.linspace(media_calculada - 4*desviacion_calculada, media_calculada + 4*desviacion_calculada, 1000)
    # Calcular la curva de densidad de probabilidad (PDF)
    pdf = norm.pdf(x, media_calculada, desviacion_calculada)

    plt.figure(figsize=(10, 6))
    plt.plot(x, pdf, 'black', linewidth=2) # Dibuja la curva

    # L贸gica para sombrear el 谩rea bajo la curva
    x_sombreado = np.linspace(x_min_sombreado, x_max_sombreado, 100)
    pdf_sombreado = norm.pdf(x_sombreado, media_calculada, desviacion_calculada)
    plt.fill_between(x_sombreado, pdf_sombreado, color='skyblue', alpha=0.6)

    # A帽adir una l铆nea en la media
    plt.axvline(media_calculada, color='red', linestyle='dashed', linewidth=2, label=f'Media = {media_calculada}')

    # T铆tulos y etiquetas
    plt.title(titulo)
    plt.xlabel('Valor')
    plt.ylabel('Densidad de Probabilidad')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# 1. Calcular la probabilidad usando scipy.stats.norm.cdf
# cdf (Cumulative Distribution Function) nos da el 谩rea acumulada desde la izquierda
valor_x = 115
probabilidad = norm.cdf(valor_x, loc=media_calculada, scale=desviacion_calculada)

print(f"La probabilidad de tener un CI inferior a {valor_x} es: {probabilidad:.4f}")
print(f"Esto corresponde a un {probabilidad*100:.2f}% de la poblaci贸n.")

# 2. Visualizar el resultado
titulo_grafico = f'Probabilidad de un CI < {valor_x} (rea = {probabilidad:.2f})'
# Sombreamos desde el valor m谩s bajo posible hasta 115
graficar_normal(media_calculada, desviacion_calculada, media_calculada - 4*desviacion_calculada, valor_x, titulo_grafico)
```

驴Cu谩l es la probabilidad de seleccionar al azar una persona con CI entre 120 y 140?

```python
# 1. Definir los l铆mites del intervalo
valor_x1 = 120
valor_x2 = 140

# 2. Calcular la probabilidad para cada l铆mite y restarlas
# Probabilidad acumulada hasta 140
prob_hasta_140 = norm.cdf(valor_x2, loc=media_calculada, scale=desviacion_calculada)
# Probabilidad acumulada hasta 120
prob_hasta_120 = norm.cdf(valor_x1, loc=media_calculada, scale=desviacion_calculada)

# La probabilidad entre los dos puntos es la resta de las 谩reas
probabilidad_entre = prob_hasta_140 - prob_hasta_120

# 3. Imprimir el resultado
print(f"La probabilidad de tener un CI entre {valor_x1} y {valor_x2} es: {probabilidad_entre:.4f}")
print(f"Esto corresponde a un {probabilidad_entre*100:.2f}% de la poblaci贸n.")

# 4. Visualizar el resultado
titulo_grafico = f'Probabilidad de un CI entre {valor_x1} y {valor_x2} (rea = {probabilidad_entre:.2f})'
# Sombreamos el 谩rea entre 120 y 140
graficar_normal(media_calculada, desviacion_calculada, valor_x1, valor_x2, titulo_grafico)
```
