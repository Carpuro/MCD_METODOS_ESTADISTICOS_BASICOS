# 📊 La Distribución Normal (Campana de Gauss)

Es el modelo estadístico más importante. Nos ayuda a entender cómo se comportan la mayoría de los fenómenos en la vida real.



Muestra que la mayoría de las cosas (alturas, notas, errores, etc.) se agrupan en torno al promedio (el centro), y los extremos son muy raros.

### Distribución Normal y su Función de Distribución Acumulada (CDF)

La **función de densidad** de la distribución normal describe qué tan probable es cada valor individual.

**Fórmula:**
$$f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{1}{2} \left( \frac{x - \mu}{\sigma} \right)^2}$$

La **función de distribución acumulada (CDF)** indica la probabilidad de que la variable aleatoria tome un valor menor o igual que un cierto número:

$$F(x) = \frac{1}{\sigma \sqrt{2\pi}} \int_{-\infty}^{x} e^{-\frac{1}{2}\left(\frac{t - \mu}{\sigma}\right)^2} dt$$

En la **normal estándar (Z)**:

$$\Phi(z) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{z} e^{-\frac{t^2}{2}} dt$$

No existe una *forma cerrada elemental* para esta integral, lo que significa que **no puede expresarse usando solo operaciones matemáticas básicas** (como suma, multiplicación o logaritmos). En su lugar, se aproxima usando la *función error*:

$$\Phi(z) = \frac{1}{2}\left[1 + \text{erf}\left(\frac{z}{\sqrt{2}}\right)\right]$$

#### Explicación intuitiva
- La **densidad** indica qué tan alta es la curva en cada punto.
- La **acumulada (CDF)** muestra **cuánta probabilidad se ha acumulado** hasta cierto valor.
- La *función error* (`erf`) es una herramienta que permite calcular esa acumulación.

#### Elementos clave
- $x$: valor observado
- $\mu$: media (centro de la curva)
- $\sigma$: desviación estándar (dispersión)
- $e$, $\pi$: constantes matemáticas
- `erf`: función especial usada para calcular probabilidades acumuladas

### Distribución Normal

La **distribución normal** muestra cómo se agrupan los datos alrededor de un valor promedio ($\mu$). Su forma es una **campana simétrica**: la mayoría de los valores están cerca del centro, y pocos se alejan mucho.

**Fórmula general:**

$$f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{1}{2} \left( \frac{x - \mu}{\sigma} \right)^2}$$

donde:
- $x$: valor observado,
- $\mu$: media (centro de la curva),
- $\sigma$: desviación estándar (qué tan dispersos están los datos),
- $e$ y $\pi$: constantes matemáticas que aseguran que el área total sea 1.

---

### Distribución Normal Estándar (Z)

Es un caso especial de la normal donde $\mu = 0$ y $\sigma = 1$. Permite comparar valores de distintas distribuciones al convertirlos en **valores Z**.

**Fórmula:**

$$f(z) = \frac{1}{\sqrt{2\pi}} e^{-\frac{z^2}{2}}$$

donde $z$ representa un valor estandarizado.

### Significado de los símbolos

- **$\mu$ (mu):** Mueve el centro de la campana (posición promedio).
- **$\sigma$ (sigma):** Controla qué tan ancha o estrecha es la campana (dispersión).
- **$e$, $\pi$:** Constantes que mantienen la forma y el total de probabilidad en 1 (100%).

## Estadísticos Asociados

> Una Distribución Normal está completamente definida por su media ($\mu$) y su desviación estándar ($\sigma$).

#### Propiedades Centrales

**Simetría:** La campana es perfectamente simétrica respecto a la media ($\mu$).

**Igualdad de Medidas:** En el centro de la curva, coinciden: Media = Mediana = Moda

## Usos Importantes

> La Distribución Normal es la "columna vertebral" de la estadística moderna.

**Modelar Fenómenos**

Se usa para representar características naturales. Ejemplos: La altura de las personas, las notas de un curso, el tiempo de reacción en un experimento.

**Control de Calidad**

Las empresas la usan para establecer tolerancias. Si un producto cae fuera de $2\sigma$ o $3\sigma$, se considera defectuoso.

**Inferencia Estadística**

Es la base para hacer predicciones. Permite a los investigadores sacar conclusiones sobre una gran población (por ejemplo, todos los votantes) estudiando solo una pequeña muestra de ella.

# Distribución de Puntuaciones de CI

El Coeficiente Intelectual (CI) es una puntuación estandarizada diseñada para que los resultados de una población grande sigan una distribución normal. Por definición, la puntuación media (μ) del CI es 100 y la desviación estándar (σ) es 15 puntos.

Vamos a simular los resultados de CI de un grupo de 1,500 personas para ver cómo se distribuyen.

$$Z = \frac{x - \mu}{\sigma}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import seaborn as sns

sns.set_style('whitegrid')
```

```python
# Parámetros de la distribución de CI
media_ci = 100
desviacion_ci = 15
n_personas = 1500

# Generamos las puntuaciones de CI simuladas
puntuaciones_ci = np.random.normal(loc=media_ci, scale=desviacion_ci, size=n_personas)

# Mostramos las primeras 10 puntuaciones generadas
print("Primeras 10 puntuaciones de CI:", puntuaciones_ci[:10])
```

```python
# Crear el histograma de las puntuaciones de CI
plt.figure(figsize=(10, 6))
plt.hist(puntuaciones_ci, bins=30, density=True, alpha=0.7, color='indigo', label='Histograma de CI')

# Añadir títulos y etiquetas
plt.title('Distribución de Puntuaciones de CI (n=1500)')
plt.xlabel('Puntuación de CI')
plt.ylabel('Densidad')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 7))

# Histograma de los datos
plt.hist(puntuaciones_ci, bins=30, density=True, alpha=0.7, color='indigo', label='Histograma de CI')

# Calculamos la media y desviación de nuestros datos para la curva
media_calculada = np.mean(puntuaciones_ci)
desviacion_calculada = np.std(puntuaciones_ci)

# Creamos el rango de valores para la curva teórica
xmin, xmax = plt.xlim()
x_axis = np.linspace(xmin, xmax, 100)
pdf = norm.pdf(x_axis, media_calculada, desviacion_calculada)

# Dibujamos la curva normal
plt.plot(x_axis, pdf, 'k', linewidth=2, color='lime', label='Curva Normal Teórica')

# Añadimos una línea para la media
plt.axvline(media_calculada, color='red', linestyle='dashed', linewidth=2, label=f'Media: {media_calculada:.2f}')

# Títulos y leyenda
plt.title('Distribución de Puntuaciones de CI con Curva Normal')
plt.xlabel('Puntuación de CI')
plt.ylabel('Densidad')
plt.legend()
plt.show()

print(f"Media calculada: {media_calculada:.2f}")
print(f"Desviación estándar calculada: {desviacion_calculada:.2f}")
```

¿Cuál es la probabilidad de seleccionar al azar a una persona con un CI inferior a 115?

```python
# Función para graficar la distribución y el área de probabilidad
def graficar_normal(media_calculada, desviacion_calculada, x_min_sombreado, x_max_sombreado, titulo):
    # Crear el rango de valores para el eje X
    x = np.linspace(media_calculada - 4*desviacion_calculada, media_calculada + 4*desviacion_calculada, 1000)
    # Calcular la curva de densidad de probabilidad (PDF)
    pdf = norm.pdf(x, media_calculada, desviacion_calculada)

    plt.figure(figsize=(10, 6))
    plt.plot(x, pdf, 'black', linewidth=2) # Dibuja la curva

    # Lógica para sombrear el área bajo la curva
    x_sombreado = np.linspace(x_min_sombreado, x_max_sombreado, 100)
    pdf_sombreado = norm.pdf(x_sombreado, media_calculada, desviacion_calculada)
    plt.fill_between(x_sombreado, pdf_sombreado, color='skyblue', alpha=0.6)

    # Añadir una línea en la media
    plt.axvline(media_calculada, color='red', linestyle='dashed', linewidth=2, label=f'Media = {media_calculada}')

    # Títulos y etiquetas
    plt.title(titulo)
    plt.xlabel('Valor')
    plt.ylabel('Densidad de Probabilidad')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# 1. Calcular la probabilidad usando scipy.stats.norm.cdf
# cdf (Cumulative Distribution Function) nos da el área acumulada desde la izquierda
valor_x = 115
probabilidad = norm.cdf(valor_x, loc=media_calculada, scale=desviacion_calculada)

print(f"La probabilidad de tener un CI inferior a {valor_x} es: {probabilidad:.4f}")
print(f"Esto corresponde a un {probabilidad*100:.2f}% de la población.")

# 2. Visualizar el resultado
titulo_grafico = f'Probabilidad de un CI < {valor_x} (Área = {probabilidad:.2f})'
# Sombreamos desde el valor más bajo posible hasta 115
graficar_normal(media_calculada, desviacion_calculada, media_calculada - 4*desviacion_calculada, valor_x, titulo_grafico)
```

¿Cuál es la probabilidad de seleccionar al azar una persona con CI entre 120 y 140?

```python
# 1. Definir los límites del intervalo
valor_x1 = 120
valor_x2 = 140

# 2. Calcular la probabilidad para cada límite y restarlas
# Probabilidad acumulada hasta 140
prob_hasta_140 = norm.cdf(valor_x2, loc=media_calculada, scale=desviacion_calculada)
# Probabilidad acumulada hasta 120
prob_hasta_120 = norm.cdf(valor_x1, loc=media_calculada, scale=desviacion_calculada)

# La probabilidad entre los dos puntos es la resta de las áreas
probabilidad_entre = prob_hasta_140 - prob_hasta_120

# 3. Imprimir el resultado
print(f"La probabilidad de tener un CI entre {valor_x1} y {valor_x2} es: {probabilidad_entre:.4f}")
print(f"Esto corresponde a un {probabilidad_entre*100:.2f}% de la población.")

# 4. Visualizar el resultado
titulo_grafico = f'Probabilidad de un CI entre {valor_x1} y {valor_x2} (Área = {probabilidad_entre:.2f})'
# Sombreamos el área entre 120 y 140
graficar_normal(media_calculada, desviacion_calculada, valor_x1, valor_x2, titulo_grafico)
```
